```python
import numba
```

```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

%timeit fib(50)
```

```python
@numba.jit
def fib(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

%timeit(fib(50))
```

```python
fib.py_func(50)
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
data = pd.DataFrame({'time': np.arange(100),
                     'temperature': 60 + np.random.randn(100).cumsum()})
```

```python
plt.plot(data['time'], data['temperature'])
plt.xlabel('time')
plt.legend(['temperature'], loc=3)
plt.ylim(0, 70);
```

```python
plt.plot('time', 'temperature', data=data)
```
